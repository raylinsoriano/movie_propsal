# Creating a new column for ROI

```python
imbd_basic['return_on_invest'] = ((imbd_basic['domestic_gross'] - imbd_basic['production_budget']) / (imbd_basic['production_budget']) * 100).copy()
```

```python
imbd_basic.head()
```

```python
#roi = list(imbd_basic['return_on_invest'])
```

```python
imbd_basic['release_date'] = pd.to_datetime(imbd_basic['release_date'])
```

```python
imbd_basic.head()
```

```python
new_roi = imbd_basic[(imbd_basic['return_on_invest'] > 33) & (imbd_basic['return_on_invest'] < 170)] 
#before I had it greater then 0 and less the 400
```

```python
new_roi.describe()
```

```python
fig, ax = plt.subplots(figsize= (10,10))
ax.hist(x = new_roi['return_on_invest'], bins = 50)
```

showing the fequncey of movie buget ROI

```python
top_ten_r = imbd_basic.nlargest(11,'return_on_invest')
```

```python
#imbd_basic['genres'].sort_values()[:100]
fig, ax = plt.subplots(figsize= (10,10))
ax.bar(x= top_ten_r['genres'], height = top_ten_r['return_on_invest'] )
plt.xticks(rotation = 'vertical')
ax.set_xlabel('Top Ten Movie Genres')
ax.set_ylabel('ROI')
ax.set_title('ROI of Top Ten Movie Genres')
```

```python
imbd_basic.head()
re_lease = imbd_basic.groupby(imbd_basic['release_date'].dt.strftime('%m'))['return_on_invest'].mean()
```

```python
release = pd.DataFrame(re_lease)
```

```python
fig, ax = plt.subplots()
ax.bar(x= release.index, height = release['return_on_invest'] )
plt.xticks() #rotation = 'vertical'
ax.set_xlabel('Release Date by Month')
ax.set_ylabel('ROI')
ax.set_title('ROI of Release Date')
```
